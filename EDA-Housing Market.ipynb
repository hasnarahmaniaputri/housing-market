```python
df = df[df["BuildingArea"]!=0]
```

```python
# menghitung outlier pada kolom Price
Q1 = df["Price"].describe()["25%"]
Q3 = df["Price"].describe()["75%"]
IQR = Q3 - Q1

batas_bawah = Q1 - (1.5 * IQR)
batas_atas = Q3 + (1.5 * IQR)
outlier_price = df[(df["Price"] < batas_bawah) | (df["Price"] > batas_atas)][["Price"]]
print(outlier_price.head())
```

```python
print(f"Total outlier pada kolom Price: {len(outlier_price)}")
```

```python
plt.style.use("seaborn")

fig, (ax) = plt.subplots(ncols=1,figsize=[26, 10])
_ = fig.suptitle("Price",size=20)
plt.subplot(222)

ax = sns.distplot(df["Price"], hist=True, rug=True, kde=True, fit=None,
                   rug_kws={"color": "red", "lw": 2, "label": "rug"},
                   kde_kws={"color": "k", "lw": 3, "label": "KDE"},
                   hist_kws={"histtype": "stepfilled", "linewidth": 1,
                             "alpha": 1, "color": "darkblue"},
                   norm_hist=True, axlabel="Price", label="Price")
ax.set(xlabel="Price", ylabel="Frequency", title="Plot Skewness pada Kolom Price")

plt.show()
```

Berdasarkan plot skewness tersebut dapat dilihat bahwa plot memiliki skewness ke kiri (skewness positif), yang artinya bahwa rata-rata persebaran data pada kolom Price mendekati nilai minimumnya, dengan total outliers sebanyak 288, sehingga dapat dikatakan bahwa kolom Price tidak berdistribusi normal.

```python
plt.figure(figsize=(10,6))
sns.barplot(x = "Type", y="Price", data=df, hue="Regionname", estimator=np.mean)
plt.title("Rata-Rata Harga Rumah Berdasarkan Tipe Rumah")
plt.show()
```

```python
tahun = df.groupby("YearBuilt").mean()
harga_tahun = tahun.reset_index()
harga_tahun

```

```python
x = harga_tahun["YearBuilt"]
y = harga_tahun["Price"]
plt.plot(x, y, "red")
plt.title('Harga Berdasarkan Tahun Pembangunan', size=15)
plt.xlabel("Year", size=15)
plt.ylabel("Price", size=15)
plt.grid()
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = pd.DatetimeIndex(df["Date"]).month
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Month", y="Price", hue = "Type")
plt.title("Tren Harga Rumah Berdasarkan Tipe Rumah")
plt.show()
```

```python
# Subplot variabel numerik terhadap Price
df.select_dtypes(["float64","int64"]).columns
```

```python
# Subplot

sns.set_style("darkgrid")
f, axes = plt.subplots(3,2, figsize = (20,30))

# Plot [0,0]
axes[0,0].scatter(x = "Rooms", y = "Price", data = df, edgecolor = "r")
axes[0,0].set_xlabel("Rooms")
axes[0,0].set_ylabel("Price")
axes[0,0].set_title("Rooms - Price")

# Plot [0,1]
axes[0,1].scatter(x = "Distance", y = "Price", data = df, edgecolor = "r")
axes[0,1].set_xlabel("Distance")
axes[0,1].set_ylabel("Price")
axes[0,1].set_title("Distance - Price")

# Plot [1,0]
axes[1,0].scatter(x = "Bathroom", y = "Price", data = df, edgecolor = "r")
axes[1,0].set_xlabel("Bathroom")
axes[1,0].set_ylabel("Price")
axes[1,0].set_title("Bathroom - Price")


# Plot [1,1]
axes[1,1].scatter(x = "Landsize", y = "Price", data = df, edgecolor = "r")
axes[1,1].set_xlabel("Landsize")
axes[1,1].set_ylabel("Price")
axes[1,1].set_title("Landsize - Price")

# Plot [2,0]
axes[2,0].scatter(x = "BuildingArea", y = "Price", data = df, edgecolor = "r")
axes[2,0].set_xlabel("BuildingArea")
axes[2,0].set_ylabel("Price")
axes[2,0].set_title("Building Area - Price")


# Plot [2,1]
axes[2,1].scatter(x = "Propertycount", y = "Price", data = df, edgecolor = "r")
axes[2,1].set_xlabel("Propertycount")
axes[2,1].set_ylabel("Price")
axes[2,1].set_title('Property Count - Price')

plt.show()
```

Berdasarkan subplot di atas dapat diketahui bahwa mayoritas rumah memiliki 3-5 kamar dan jumlah kamar mandi sebanyak 2-4 unit.
Terdapat korelasi negatif antara jarak dari pusat distrik dengan harga rumah. Semakin dekat jarak rumah ke kawasan pusat bisnis, maka harga rumah semakin mahal.

```python
# Korelasi variabel
dfcorr = df[["Price","Rooms","Distance","Bathroom","Landsize","BuildingArea","Propertycount"]]
dfcorr.head()
dfcorr.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(dfcorr.corr(),cmap = 'coolwarm',linewidth = 1,annot= True, annot_kws={"size": 9})
plt.title("Korelasi Variabel")
```

Berdasarkan tabel korelasi dan plot heatmap di atas dapat diketahui bahwa terdapat korelasi positif yang kuat antara Price dengan Rooms dan Building Area, dengan nilai korelasi tertinggi yaitu sebesar 0,53.
Selanjutnya, Bathroom juga memiliki korelasi positif terhadap Price dengan nilai korelasi sebesar 0,5.

Variabel Distance dan Propertycount berkorelasi negatif dengan variabel Price, sementara variabel Landsize memiliki korelasi yang sangat lemah dengan variabel Price.

Hal ini dapat diartikan bahwa **semakin banyak jumlah kamar dan semakin besar luas bangunannya, maka harga jual rumah tersebut semakin mahal.**
